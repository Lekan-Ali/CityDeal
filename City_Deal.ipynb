## 2. Applying data cleaning and Validation

```python
## CODE 9 ##

# create a copy of the combined raw dataset (this is the dataset, we will be using going forward)
df1 = df_outcomes_activities_demographic.copy()
```

```python
## CODE 10 ##

# count the number of missing observations in each feature (and sort it)
df1.isna().sum().sort_values(ascending = False)
```

```python
## CODE 11 ##

# calculate the percentage of missing values


# create a new variable containing the count of missing observations and use reset_index to convert to df
df1_missing_observations = df1.isna().sum().sort_values(ascending = False).reset_index()


# rename the headers
df1_missing_observations.rename(columns={'index':'Features',0:'Missing_entries'}, inplace = True)


# create a new feature or column that contains the percentage share of missing values
df1_missing_observations['percentage_of_total_entries'] = round((df1_missing_observations['Missing_entries']/1697592)*100,2)
```

```python
# Display the first 30 features with their total number of missing entries and percentage share of missing entries
df1_missing_observations[1:30]
```

```python
## CODE 12 ##

# group features into band such that features that are missing more than 95% are in a single group, same for others
# The aim is to visualize the groups


# creating a function for the grouping
def missing_entries_band(df):
    
    if (df['percentage_of_total_entries']) > 95:
        return '> 95%'
    elif (df['percentage_of_total_entries'] <= 95) and (df['percentage_of_total_entries'] > 90):
        return '91% - 95%'
    elif (df['percentage_of_total_entries'] <= 90) and (df['percentage_of_total_entries'] > 80):
        return '81% - 90%'
    elif (df['percentage_of_total_entries'] <= 80) and (df['percentage_of_total_entries'] > 70):
        return '71% - 80%'
    elif (df['percentage_of_total_entries'] <= 70) and (df['percentage_of_total_entries'] > 60):
        return '61% - 70%'
    elif (df['percentage_of_total_entries'] <= 60) and (df['percentage_of_total_entries'] > 50):
        return '51% - 60%'
    elif (df['percentage_of_total_entries'] <= 50) and (df['percentage_of_total_entries'] > 40):
        return '41% - 50%'
    elif (df['percentage_of_total_entries'] <= 40) and (df['percentage_of_total_entries'] > 30):
        return '31% - 40%'
    elif (df['percentage_of_total_entries'] <= 30) and (df['percentage_of_total_entries'] > 20):
        return '21% - 30%'
    elif (df['percentage_of_total_entries'] <= 20) and (df['percentage_of_total_entries'] > 10):
        return '11% - 20%'
    elif (df['percentage_of_total_entries'] <= 10):
        return '<= 10%'
    
    

# create a new column in the dataframe and use the apply method to assign each feature to a group
df1_missing_observations['groups'] = df1_missing_observations.apply(missing_entries_band, axis = 1)
```

```python
df1_missing_observations
```

```python
## CODE 13 ##


# Visualize the percentage of missing values


# create a new variable
df1_missing_band = df1_missing_observations['groups'].value_counts().reset_index()


# rename the headers
df1_missing_band.rename(columns={'index':'groups','groups':'count_of_features'}, inplace = True)

# re-arrange index
#df1_missing_band = df1_missing_band.reindex([1,9,7,4,5,8,10,6,3,2,0])
df1_missing_band = df1_missing_band.reindex([0,2,3,6,10,8,5,4,7,9,1])
```

```python
df1_missing_band
```

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.set_title("This is a title")

ax.grid(linewidth=.25, zorder = 1, which = 'major', axis = 'x'
        , ls='--')

#ax.barh()

ax.barh(y = 'groups', width = 'count_of_features', data = df1_missing_band);
```

```python
## CODE 14 ##

# drop features with more than 100%

# filter for features that are missing all observations or have 100% percentage of missing values
to_be_dropped = df1_missing_observations[df1_missing_observations['percentage_of_total_entries'] == 100]['Features']
```

```python
# create a loop to remove features that have 100% missing entries

# takes too long to execute, doesn't make sense. 
#df2 = []

#for i in to_be_dropped:
#    j = df1.drop(i, axis = 1)
#    df2.append(j)
```

```python
# dropping features that are missing all entries and assigning to a new variable df2
df2 = df1.drop(to_be_dropped, axis = 1)
```
